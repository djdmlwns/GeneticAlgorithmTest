```python
import numpy as np
x = np.linspace(0,2*np.pi,1000)
y = np.sin(x) + np.random.random(1000) * 0.4
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
```

```python
from scipy.signal import savgol_filter
yhat = savgol_filter(y, 11, 2) # window size 51, polynomial order 3
```

```python
plt.plot(x,yhat)
plt.plot(x,y)
```

```python
def movingaverage(x,y,windowsize):
    yhat = []
    if len(x) < windowsize:
        raise("Window size must be larger than the data point")
    else:
        for i in range(len(x)):
            iteration = 0
            yhat_element = 0
            while iteration <= windowsize:
                if i >= iteration:
                    yhat_element += y[i - iteration]
                    iteration += 1
                else:
                    break
            yhat.append(yhat_element / (iteration))
    plt.plot(x,y)
    plt.plot(x,yhat)

    return yhat
                
                
           
            
```

```python
yhat = movingaverage(x,y,30)
```
